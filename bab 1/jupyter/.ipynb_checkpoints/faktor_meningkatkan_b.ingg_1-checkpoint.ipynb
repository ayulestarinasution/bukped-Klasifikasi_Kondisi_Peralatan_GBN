```python
data = df.copy()
```

```python
target = 'Kemampuan_Ingg'
```

```python
encode = ['Nama']
```

```python
data.head()
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target = {'Mampu':0, 'tidak Mampu':1}
```

```python
def target_encode(val):
    return target[val]
```

```python
data['Kemampuan_Ingg'] = data['Kemampuan_Ingg'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('Kemampuan_Ingg', axis=1)
Y = data['Kemampuan_Ingg']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=3)

print('Train set :', X_train.shape, y_train.shape)
print('Test set :', X_test.shape, y_test.shape)
```

```python
Y
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier

model_dtree = DecisionTreeClassifier(max_depth=3, criterion="entropy")
model_dtree.fit(X_train, y_train)
```

```python
y_pred = model_dtree.predict(X_test)
```

```python
print(y_pred[0:5])
print(y_test[0:5])
```

```python
from sklearn.metrics import accuracy_score

print('Akurasi Decision Tree :', accuracy_score(y_test, y_pred))
```
